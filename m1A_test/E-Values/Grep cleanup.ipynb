```python
newFile=''
with open("eValues.txt") as file:
    for n,line in enumerate( file.readlines() ):
        if n < 213:
            thing = line.split(' ')
            #print( thing[len(thing)-1] )
            newFile+=thing[len(thing)-1]
        else:
            newFile+=line
print(newFile)    
      
with open("eValues.txt", 'w') as file:
    file.write(newFile)
```
