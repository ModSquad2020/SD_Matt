# Converting eValue grep into just numbers
Need to know when the inconsitency starts

```python
newFile=''
with open("eValues.txt") as file:
    for n,line in enumerate( file.readlines() ):
        if n < 258:
            thing = line.split(' ')
            #print( thing[len(thing)-1] )
            newFile+=thing[len(thing)-1]
        else:
            newFile+=line
print(newFile, newFile.count('\n'))    
    
with open("eValues.txt", 'w') as file:
    file.write(newFile)
```

```python
newFile=""
with open("tRNAnames.txt") as file:
    for n,line in enumerate(file.readlines()):
        #print(n, line.split(":")[1].split(' ')[0])
        newFile= newFile+ '{}\n'.format(line.split(":")[1].split(' ')[0])
print(newFile) 

with open('tRNAnames.txt', 'w') as file:
    file.write(newFile)
```

# Making the file for Henry's graphing tool
Currently printing out to the for Henry.txt file

```python
import pandas as pd
import sys
sys.path.insert(1, '../../../python scripts/todd stuff/BioTools')
from sequenceAnalysis import FastAreader
gucciDic = {}
#neg + creating dic
with open('eValues_neg.txt') as file1:
    with open('tRNAnames_neg.txt') as file2:
        for name, evalue in zip(file2, file1):
            name=name.replace('\n', '')
            evalue=float( evalue.replace('\n', '') )
            gucciDic[name] = [evalue]


#pos + appending to dic            
with open('eValues_pos.txt') as file1:
    with open('tRNAnames_pos.txt') as file2:
        for name, evalue in zip(file2, file1):
            name=name.replace('\n', '')
            evalue=float( evalue.replace('\n', '') )
            try:
                gucciDic[name].append(evalue)
                gucciDic[name] = gucciDic[name]
            except KeyError:
                gucciDic[name] = [1, evalue]

#correcting the dictionary (adding 1 for the +)             
for key, value in gucciDic.items():
    if len(value) == 1:
        gucciDic[key].append(1)
        gucciDic[key]=gucciDic[key]
        
#getting the + or - for coloring
for header, sequence in FastAreader('hasM1A.fa').readFasta():
    header = header.split(' ')[0]
    for key, value in gucciDic.items():
        if key == header:
            gucciDic[key].append('+')
            break

#getting the - for coloring
for key, value in gucciDic.items():
    if len(value) == 2:
        gucciDic[key].append('-')

print(gucciDic)
df = pd.DataFrame.from_dict(gucciDic, orient='index')
df = df[[2,1,0]]
#df.to_csv('forHenry.txt')
```
