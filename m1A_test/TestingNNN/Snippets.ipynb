```python
hasMod = ["Arg-TCT", "Arg-ACG",
          "Asn-GTT", "Cys-GCA",
          "Ile-AAT", "Ile-TAT",
          "Leu-TAG", "Leu-TAA",
          "Lys-CTT", "Lys-TTT",
          "Met-CAT", "Phe-GAA",
          "Pro-TGG", "Ser-GCT",
          "Thr-AGT", "Trp-CCA",
          "Val-TAC", "Val-CAC",
          "Val-AAC", "iMet-CAT"
]
```

```python
reader = FastAreader("AlignedUsingEuk.fa")
with open('hasM1A.fa', 'w') as file: 
    for header,seq in reader.readFasta():
        tRNA = f'>{header}\n{NNNSlicer.threeRegions(seq, "AA-acceptor1", "AA-acceptor2", "T-arm")}\n'
        for thing in hasMod:
            if thing in header:
                print(tRNA)
                file.write(tRNA)
                
```

```python
noDashes=""
with open("hasM1A.fa") as file:
    noDashes=file.read()
    noDashes=noDashes.replace("-","")
    print(noDashes)
with open("hasM1A_NoDash.fa", "w") as file:
    file.write(noDashes)
```
