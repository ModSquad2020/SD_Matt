```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

df = pd.read_csv('TESTresultsBetter.csv', delimiter=',')
df2 = pd.read_csv('distilledModomicsBetter.csv', delimiter=',')
```

# Working with TESTresults
This will focus on splitting up columns into several columns

## Splits Query CM
This section splits the Query CM column by the "\[, \], \_"
keys. Then renames the new splits accordingly.

```python
df[['Query CM Header', "Query CM Arms"]] = df['Query CM'].str.split('[',expand=True)
df["Query CM Arms"] = df['Query CM Arms'].str.split(']',expand=True)
df[['Against Organism', "Kingdom CM", "Modification", "Position", "CM Charge", "Empty Column"]] = df['Query CM Header'].str.split('_',expand=True)
```

## Cleaning up columns
This section removes the old "Query CM" and intermidiary
"Query CM Header"/"Empyt Column" columns that pop up. 

```python
#df.drop(['Query CM Header','Empty Column', "Query CM"], axis=1, inplace=True)
df.drop(['Query CM Header','Empty Column'], axis=1, inplace=True)
df.head()
```
